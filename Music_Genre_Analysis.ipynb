```python
from collections import defaultdict 
import csv 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

def genre_attributes(artist_data, genre):
    attribute_dict = defaultdict(list)
    for row in artist_data:
        if row[2] == genre:
            attribute_dict["danceability"].append(row[4])
            attribute_dict["energy"].append(row[5])
            attribute_dict["valence"].append(row[6])
            attribute_dict["tempo"].append(row[7])
            attribute_dict["loudness"].append(row[8])
            attribute_dict["key"].append(row[9])
            attribute_dict["instrumentalness"].append(row[10])
            attribute_dict["liveness"].append(row[11])
            attribute_dict["speechiness"].append(row[12])
            attribute_dict["duration"].append(row[13])
    return attribute_dict
```

```python
# As a csv.reader is an expendible resource we write the lines of our csv file to a list
with open("data_by_artist_with_genre.csv", "rt") as infile:
        reader = csv.reader(infile)
        # We discard the column labels as they are unneccesary 
        next(reader)
        artist_data = list(reader)
```

```python
artist_df = pd.read_csv("data_by_artist_with_genre.csv")
genres = pd.unique(artist_df["artist_genre"])
print("We have " + str(len(genres)) + " genres.")
print(artist_df["artist_genre"].value_counts())
```

```python
def music_attrib_boxplots(artist_data, genre):
    attributes_dict = genre_attributes(artist_data, genre)
    att_data_frame = pd.DataFrame(data = attributes_dict)
    f, axes = plt.subplots(2, 5, figsize=(24, 10), sharex=True)
    ax1 = sns.boxplot(data=att_data_frame["danceability"], ax=axes[0, 0])
    ax1 = sns.stripplot(data=att_data_frame["danceability"], ax=axes[0, 0], color = "0.3")
    ax1.set(xlabel = "danceability")
    ax2 = sns.boxplot(data=att_data_frame["energy"],ax=axes[0, 1])
    ax2 = sns.stripplot(data=att_data_frame["energy"], ax=axes[0, 1], color = "0.3")
    ax2.set(xlabel = "energy")
    ax3 = sns.boxplot(data=att_data_frame["valence"],ax=axes[1, 0])
    ax3 = sns.stripplot(data=att_data_frame["valence"], ax=axes[1, 0], color = "0.3")
    ax3.set(xlabel = "valence")
    ax4 = sns.boxplot(data=att_data_frame["tempo"],ax=axes[1, 1])
    ax4 = sns.stripplot(data=att_data_frame["tempo"], ax=axes[1, 1], color = "0.3")
    ax4.set(xlabel = "tempo")
    ax5 = sns.boxplot(data=att_data_frame["loudness"],ax=axes[1, 2])
    ax5 = sns.stripplot(data=att_data_frame["loudness"], ax=axes[1, 2], color = "0.3")
    ax5.set(xlabel = "loudness")
    ax5.set(ylabel = "dB")
    ax6 = sns.boxplot(data=att_data_frame["key"],ax=axes[1, 3])
    ax6 = sns.stripplot(data=att_data_frame["key"], ax=axes[1, 3], color = "0.3")
    ax6.set(xlabel = "key")
    ax7 = sns.boxplot(data=att_data_frame["instrumentalness"],ax=axes[1, 4])
    ax7 = sns.stripplot(data=att_data_frame["instrumentalness"], ax=axes[1, 4], color = "0.3")
    ax7.set(xlabel = "instrumentalness")
    ax8 = sns.boxplot(data=att_data_frame["liveness"],ax=axes[0, 2])
    ax8 = sns.stripplot(data=att_data_frame["liveness"], ax=axes[0, 2], color = "0.3")
    ax8.set(xlabel = "liveness")
    ax9 = sns.boxplot(data=att_data_frame["speechiness"],ax=axes[0, 3])
    ax9 = sns.stripplot(data=att_data_frame["speechiness"], ax=axes[0, 3], color = "0.3")
    ax9.set(xlabel = "speechiness")
    ax10 = sns.boxplot(data=att_data_frame["duration"],ax=axes[0, 4])
    ax10 = sns.stripplot(data=att_data_frame["duration"], ax=axes[0, 4], color = "0.3")
    ax10.set(xlabel = "duration")
    ax10.set(ylabel = "ms")
    plt.show()
```
